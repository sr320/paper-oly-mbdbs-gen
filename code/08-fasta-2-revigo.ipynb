### Blast gene fasta files for DMLs, MACAU and their background files against uniprot database.

Uniprot/Swiss-prot reviewed database downloaded on 02/08/2020 from https://www.uniprot.org/downloads

```python
bldir = "../resources/ncbi-blast-2.7.1+/bin/"
```

```python
! {bldir}blastx -help
```

```python
! {bldir}makeblastdb \
-in ../resources/uniprot_sprot.fasta \
-dbtype prot \
-out ../resources/uniprot_sprot_20200208.fasta
```

```python
! {bldir}blastx \
-query ../analyses/Olurida_v081.DML-gene.fa \
-db ../resources/uniprot_sprot_20200208.fasta \
-num_threads 4 \
-evalue 1E-5 \
-outfmt 6 \
-max_target_seqs 1 \
-out ../analyses/DML-gene_blastx_sp.tab
```

```python
! {bldir}blastx \
-query ../analyses/Olurida_v081.AllLociDMLs-gene.fa \
-db ../resources/uniprot_sprot_20200208.fasta \
-num_threads 4 \
-evalue 1E-5 \
-outfmt 6 \
-max_target_seqs 1 \
-out ../analyses/AllLociDMLs-gene_blastx_sp.tab
```

```python
! {bldir}blastx \
-query ../analyses/Olurida_v081.macau-gene.fa \
-db ../resources/uniprot_sprot_20200208.fasta \
-num_threads 4 \
-evalue 1E-5 \
-outfmt 6 \
-max_target_seqs 1 \
-out ../analyses/macau-gene_blastx_sp.tab
```

```python
! {bldir}blastx \
-query ../analyses/Olurida_v081.AllLociMACAU-gene.fa \
-db ../resources/uniprot_sprot_20200208.fasta \
-num_threads 4 \
-evalue 1E-5 \
-outfmt 6 \
-max_target_seqs 1 \
-out ../analyses/AllLociMACAU-gene_blastx_sp.tab
```
